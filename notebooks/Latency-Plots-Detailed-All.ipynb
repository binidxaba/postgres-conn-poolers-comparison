# Analyze Latencies

In this notebook we explore the latency of the connection poolers for long running experiments. The input are csv files containing percentile information calculated for every second in the experiment.


```python
import pandas as pd
import matplotlib.pyplot as plt
```

These are the files that we read to generate the plots. You should change these if you get new results.

```python
directories = [
    "./002-results-detailed-latencies/pgbouncer-1707335134",
    "./002-results-detailed-latencies/pgcat-1707322429",
    "./002-results-detailed-latencies/supavisor-1707339359"
]
    
configs = [
    "(PgBouncer, clients = 1250)",
    "(PgCat, clients = 1250)",
    "(Supavisor, clients = 1250)"
]
```

These are just constants. No need to change them.

```python
colors = ["#412e93", "#f9457a", "#da38ca", "#44d3ec", "#e9fc87",
            "#1f78b4",  # Blue
            "#33a02c",  # Green
            "#e31a1c",  # Red
            "#ff7f00",  # Orange
            "#6a3d9a",  # Purple
            "#a6cee3",  # Light Blue
            "#b2df8a",  # Light Green
            "#fb9a99",  # Light Red
            "#fdbf6f",  # Light Orange
            "#cab2d6"   # Light Purple
        ]
figsize = (16,9)
```

## Latency for p50

```python
dfs = []
for d in directories:
    dfs.append(pd.read_csv(d + '/q50.csv'))#.set_index('time_epoch'))
```

```python
title = f"Query Latency (p50)"
max_y = dfs[2]['time'].max() * 1.5
for i, dp in enumerate(dfs):
    if i == 0:
        ax = dp.plot.line(y='time', figsize=figsize, ylim=(0,max_y),title=title, label=configs[i], color=colors)
    else:
        ax = dp.plot.line(ax=ax, y='time', label=configs[i])

ax.set_ylabel('Latency (ms)')
ax.set_xlabel('Sample')
```
